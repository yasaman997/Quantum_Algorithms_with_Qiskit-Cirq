```python
qc=QuantumCircuit(4)
#qc.ccx is the tofolli with the two first qubit as the control, and last one as target
#The Toffoli gate, also known as the double controlled-NOT gate (CCX), has two control qubits and one target. 
#It applies a NOT to the target only when both controls are in state ∣1⟩.
# The Toffoli gate with the Hadamard gate is a universal gate set for quantum computing.
qc.ccx(0,1,2)
qc.ccx(0,1,2)

#ch operator, control on hadamard, if qubit 0, was 1, act hadamard on 1 qubit
qc.ch(0,1)
# control phase operator, first determine rotation parameter, then control qubit, then target qubit
qc.cp(math.pi/4,0,1)
# control rx, rotation on x axis first determine rotation parameter, then control qubit, then target qubit
qc.crx(math.pi/2,2,3)
#control ry, rotation on y axis first determine rotation parameter, then control qubit, then target qubit

qc.cry(math.pi/8,2,3)

#control rz, rotation on z axis first determine rotation parameter, then control qubit, then target qubit
qc.crz(math.pi/4,0,1)

#operator that swape two qubits ( here swapping qubit 2, with qubit 3)

qc.swap(2,3)

#operator c.swap means control swap, when control qubit( here qubit 0)  was 1, swape the other two target qubit

qc.cswap(0,2,3)

# csx,it's like sx operator, however the control is first argument, the 2nd one is target
qc.csx(0,1)

#control universal,4 argument for parameters of angle and additional global phase,then control qubit, then target qubit

qc.cu(math.pi/2, 0, math.pi, 0,0,1)

qc.cx(2,3)
qc.cy(2,3)
qc.cz(1,2)

qc.draw('mpl')
```

```python
#more multi qubit gates:
#multi control phase:( phase argument, then list of control qubits, then target qubits)
qc.mcp(math.pi/4, [0,1,2],3)


qc.draw('mpl')

#the final blue ones are the ones we now added by this order, why all controls and target are looking the same?
# because the z operator has phase kick back, there is no difference between control and target qubits 
# for phase operators, the control and target opeerators are shown the same as such,because they both effect eachother
#also because of phase kick back, the target and control. looks the same on map
```

```python
#mcp has multi control qubits
qc.mcx([0,1,2],3)
qc.draw('mpl')
#for this one, target qubitis shown differently
```

```python
#how to apply barrier?
qc=QuantumCircuit(2)
#apply hadamard on more than one qubit
qc.h([0,1])
#creating barrier with code below, if no argument mentioned, it will be applied to whole circuit
qc.barrier()
qc.x(0)
qc.x(0)
qc.s(1)
# we can specify on which one barrier shows up, by specifying the argument
qc.barrier([1])
qc.s(1)
qc.draw('mpl')
```

```python
#Parametric gates used in Algorithms
from qiskit.circuit import QuantumCircuit, Parameter # import class of parameters
theta1 =Parameter('theta1')
theta2 =Parameter('theta2')
theta3 =Parameter('theta3')

qc=QuantumCircuit(3)
qc.h([0,1,2])
#for phase gate, we wanna have phase as a parameter rather than a constant number
qc.p(theta1, 0)
qc.p(theta2, 1)
qc.p(theta3, 2)
qc.draw('mpl')
```

```python
#now we have parameteric gate
```

```python
#Assigned parameters or bined parameters: assigning a number to the parameter
b_qc=qc.assign_parameters({
  theta1: math.pi/8,
  theta2: math.pi/4,
  theta3: math.pi/2,
})
b_qc.draw('mpl')
```

```python
#this method would be beneficial for simulation
```

```python
#Combining multiple circuits( without append command)
qc=QuantumCircuit(2,2) #first argument is qubit, 2nd one is bit
qc.h(0)
print("first quantum circuit")
display(qc.draw('mpl'))

#aplying cnot to the next quantum circuit
another_qc=QuantumCircuit(2,2)
another_qc.cx(0,1)
print("second quantum circuit")
display(another_qc.draw('mpl'))
#it givess bell circuit (for intangelment)

bell_qc=qc.compose(another_qc) # I have qc, I put compose then build another qc on top of it
print("composed quantum circuit")
bell_qc.draw('mpl')
```

```python
#difference between compose and append: append showed the added one in a box, compose showed it all for added one
```

```python
#we have pieces of circuits that wanna be added multiple time and wanna make a copy of it to be used multiple times
qc=QuantumCircuit(3)
qc.h([0,1,2])
new_qc=qc.copy()
display(qc.draw('mpl'))
new_qc.draw('mpl')
```

```python
#the qc and new qc are the same
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.s(0)
qc.x(1)
print("the original quantum circuit")
display(qc.draw('mpl'))

decompose_qc = qc.decompose()
print("the decomposed circuit")
decompose_qc.draw('mpl')
```

```python
# the decompose command, decomposed circuits to fundamental or universal gate, this is for compiling the circuit
#when it's gonna be applied to real circuit, this should be built of other gates
#decompose happens when do the compilling, or in simulation
```

```python
#wanna define a new gate, or compound gate (Oracle):
#we have anti CNOT, it's like CNOT, but when control is zero, it acts on target, qiskit doesn't have it, so I make it
anti_cnot_qc=QuantumCircuit(2)
anti_cnot_qc.x(0)
anti_cnot_qc.cx(0,1) #control NOT
anti_cnot_qc.x(0)
display(anti_cnot_qc.draw('mpl'))

anti_cnot_gate=anti_cnot_qc.to_gate()  # the .to_gate() command could make it a gate

qc=QuantumCircuit(3)
qc.x([0,1,2])
qc.append(anti_cnot_gate, [0,2])#since anti_cnot_qc has two qubits,it could only act on two qubits specified in list
display(qc.draw('mpl'))
qc.decompose().draw('mpl')
```
